```python
# recommendation will be based on these features only
# data = data.loc[:,['director_name','actor_1_name','actor_2_name','actor_3_name','genres','movie_title']]
# Sá»­ dá»¥ng nhiá»u Ä‘áº·c trÆ°ng hÆ¡n
features_for_recommendation = [
    'title',           # TÃªn phim
    'genres',          # Thá»ƒ loáº¡i  
    'overview',        # MÃ´ táº£ phim
    'original_language', # NgÃ´n ngá»¯
    'keywords',        # Tá»« khÃ³a (náº¿u cÃ³)
    'production_companies', # CÃ´ng ty sáº£n xuáº¥t
    'production_countries'  # Quá»‘c gia sáº£n xuáº¥t
]

# Lá»c chá»‰ cÃ¡c cá»™t tá»“n táº¡i
existing_features = [col for col in features_for_recommendation if col in data.columns]
data_for_recommendation = data.loc[:, existing_features]

print("CÃ¡c Ä‘áº·c trÆ°ng Ä‘Æ°á»£c chá»n:")
print(existing_features)
```

```python
data.head(10)
```

```python
# data['actor_1_name'] = data['actor_1_name'].replace(np.nan, 'unknown')
# data['actor_2_name'] = data['actor_2_name'].replace(np.nan, 'unknown')
# data['actor_3_name'] = data['actor_3_name'].replace(np.nan, 'unknown')
# data['director_name'] = data['director_name'].replace(np.nan, 'unknown')
data = data.fillna({
    'overview': 'unknown',
    'genres': 'unknown', 
    'tagline': 'unknown',
    'keywords': 'unknown',
    'production_companies': 'unknown',
    'budget': 0,
    'revenue': 0,
    'runtime': 0
})
```

```python
data
```

```python
# data['genres'] = data['genres'].str.replace('|', ' ')
# Xem format genres hiá»‡n táº¡i
# Xá»­ lÃ½ genres cho Content-Based Filtering
data['genres'] = data['genres'].str.replace(', ', ' ')

# Xem káº¿t quáº£
print("Genres sau khi xá»­ lÃ½:")
print(data['genres'].head(10))
```

```python
data
```

```python
# data['movie_title'] = data['movie_title'].str.lower()
# Chuyá»ƒn tÃªn phim thÃ nh lowercase
data['title'] = data['title'].str.lower()
```

```python
# null terminating char at the end
# data['movie_title'][1]
data['title'][1]
```

```python
# removing the null terminating char at the end
# data['movie_title'] = data['movie_title'].apply(lambda x : x[:-1])
# LÃ m sáº¡ch toÃ n diá»‡n hÆ¡n
data['title'] = data['title'].str.strip()  # Bá» space Ä‘áº§u/cuá»‘i
data['title'] = data['title'].str.replace('\n', '')  # Bá» xuá»‘ng dÃ²ng  
data['title'] = data['title'].str.replace('\r', '')  # Bá» carriage return
data['title'] = data['title'].str.replace('\t', '')  # Bá» tab

# Kiá»ƒm tra káº¿t quáº£
print("Titles sau khi lÃ m sáº¡ch toÃ n diá»‡n:")
print(data['title'].head(10))
```
