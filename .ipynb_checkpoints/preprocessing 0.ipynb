```python
data = pd.read_csv('D:/MovieRecommendationSystem/datasets/TMDB_movie_dataset_v11.csv')
```

```python
# we have movies only upto 2016
import matplotlib.pyplot as plt
# data.title_year.value_counts(dropna=False).sort_index().plot(kind='barh',figsize=(15,16))
# plt.show()
# Trích xuất năm từ cột release_date
data['year'] = pd.to_datetime(data['release_date']).dt.year

# Vẽ biểu đồ số lượng phim theo năm
data.year.value_counts(dropna=False).sort_index().plot(kind='barh', figsize=(15,16))
plt.title('Số lượng phim theo năm')
plt.xlabel('Số lượng phim')
plt.ylabel('Năm')
plt.show()
```

```python
# recommendation will be based on these features only
# data = data.loc[:,['director_name','actor_1_name','actor_2_name','actor_3_name','genres','movie_title']]
# Sử dụng nhiều đặc trưng hơn
features_for_recommendation = [
    'title',           # Tên phim
    'genres',          # Thể loại  
    'overview',        # Mô tả phim
    'original_language', # Ngôn ngữ
    'keywords',        # Từ khóa (nếu có)
    'production_companies', # Công ty sản xuất
    'production_countries'  # Quốc gia sản xuất
]

# Lọc chỉ các cột tồn tại
existing_features = [col for col in features_for_recommendation if col in data.columns]
data_for_recommendation = data.loc[:, existing_features]

print("Các đặc trưng được chọn:")
print(existing_features)
```

```python
data.head(10)
```

```python
# data['actor_1_name'] = data['actor_1_name'].replace(np.nan, 'unknown')
# data['actor_2_name'] = data['actor_2_name'].replace(np.nan, 'unknown')
# data['actor_3_name'] = data['actor_3_name'].replace(np.nan, 'unknown')
# data['director_name'] = data['director_name'].replace(np.nan, 'unknown')
data = data.fillna({
    'overview': 'unknown',
    'genres': 'unknown', 
    'tagline': 'unknown',
    'keywords': 'unknown',
    'production_companies': 'unknown',
    'budget': 0,
    'revenue': 0,
    'runtime': 0
})
```

```python
data
```

```python
# data['genres'] = data['genres'].str.replace('|', ' ')
# Xem format genres hiện tại
# Xử lý genres cho Content-Based Filtering
data['genres'] = data['genres'].str.replace(', ', ' ')

# Xem kết quả
print("Genres sau khi xử lý:")
print(data['genres'].head(10))
```

```python
data
```

```python
# data['movie_title'] = data['movie_title'].str.lower()
# Chuyển tên phim thành lowercase
data['title'] = data['title'].str.lower()
```

```python
# null terminating char at the end
# data['movie_title'][1]
data['title'][1]
```

```python
# removing the null terminating char at the end
# data['movie_title'] = data['movie_title'].apply(lambda x : x[:-1])
# Làm sạch toàn diện hơn
data['title'] = data['title'].str.strip()  # Bỏ space đầu/cuối
data['title'] = data['title'].str.replace('\n', '')  # Bỏ xuống dòng  
data['title'] = data['title'].str.replace('\r', '')  # Bỏ carriage return
data['title'] = data['title'].str.replace('\t', '')  # Bỏ tab

# Kiểm tra kết quả
print("Titles sau khi làm sạch toàn diện:")
print(data['title'].head(10))
```
